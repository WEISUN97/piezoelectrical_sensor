```python
from chip.cell_mesh import cell_mesh
from chip.cell_beam import cell_beam
from chip.cell_resonator import cell_resonator
from chip.cell_resonator_stiff import cell_resonator_stiff
from chip.cell_membrane import cell_membrane
from chip.cell_mesh_large import cell_mesh_large
from chip.cell_membrane_large import cell_membrane_large
from chip.cell_sides.cell_cantilever_resistive import cell_cantilever_resistive
from chip.cell_sides.cell_d_clamped_resistive import cell_d_clamped_resistive
from chip.cell_sides.cell_d_clamped_tunable import cell_d_clamped_tunable
from chip.cell_sides.cell_mesh_resistive import cell_mesh_resistive

wafer = gf.Component()

# layout of different cells on wafer

# cell_mesh
wafer << cell_mesh
wafer.add_ref(cell_mesh).movex(20000).movey(-20000)
wafer.add_ref(cell_mesh).movey(-20000)

# cell_resonator
wafer.add_ref(cell_resonator).movex(-20000)
wafer.add_ref(cell_resonator).movex(-20000).movey(-20000)
wafer.add_ref(cell_resonator).movex(-40000).movey(-20000)

# cell_resonator_stiff
wafer.add_ref(cell_resonator_stiff).movex(-40000)
wafer.add_ref(cell_resonator_stiff).movex(-40000).movey(20000)

# cell_beam
wafer.add_ref(cell_beam).movex(-20000).movey(20000)

# cell_membrane
wafer.add_ref(cell_membrane).movey(20000)

# cell_mesh_large
wafer.add_ref(cell_mesh_large).movex(20000).movey(20000)

# cell_membrane_large
wafer.add_ref(cell_membrane_large).movex(20000)

# cell_cantilever_resistive
wafer.add_ref(cell_cantilever_resistive).movex(-20000).movey(20000)

# cell_d_clamped_resistive
wafer.add_ref(cell_d_clamped_resistive).movey(20000)

# cell_d_clamped_tunable
wafer.add_ref(cell_d_clamped_tunable).movex(-20000).movey(-35830)

# cell_mesh_resistive
wafer.add_ref(cell_mesh_resistive).movex(0).movey(-35830)

# wafer.show()
```

```python
# grid
width = 750/2
circle = gf.components.circle(50000, layer=(10, 0))
circle_ref = wafer << circle
# horizontal lines of the grid
h_lines = [2, 8, 8, 8, 8, 8, 8, 8, 2]
x_size = [2, 1, 1, 1, 1, 1, 1, 1, 2]
y_space = [0, 15830, 10000, 10000, 10000, 10000, 10000, 10000, 15830]
x_temp = [-4, -3, -2, -1, 0, 1, 2, 3]
x_space = [[-2, 0], x_temp, x_temp, x_temp, x_temp, x_temp, x_temp, x_temp, [-2, 0]]
gap = 600 # um
y_space_temp = 0
for i in range(len(h_lines)):
    y_space_temp += y_space[i]
    for j in range(h_lines[i]):
        horizontal_line = gf.components.rectangle(size=(10000*x_size[i]-2*gap, width), layer=(3, 0))
        frame = wafer << horizontal_line
        frame.move((x_space[i][j]*10000+gap, 45830-y_space_temp-width/2))
# vertical lines of the grid
v_lines_len = [[10000] * 6, [10000] * 6, [15830, 10000, 10000, 10000, 10000, 10000, 10000, 15830], [10000] * 6, [15830, 10000, 10000, 10000, 10000, 10000, 10000,15830], [10000] * 6, [15830, 10000, 10000, 10000, 10000, 10000, 10000, 15830], [10000] * 6, [10000] * 6]
y_top = [20000, 20000, 30000, 20000, 30000, 20000, 30000, 20000, 20000]

for i in range(len(v_lines_len)):
    y_space_temp = y_top[i]
    for j in range(len(v_lines_len[i])):
        x_pos = -20000 + (i-2)*10000 - width/2
        vertical_line = gf.components.rectangle(size=(width, v_lines_len[i][j]-2*gap), layer=(3, 0))
        frame = wafer << vertical_line
        frame.move((x_pos,y_space_temp+gap))
        if j != len(v_lines_len[i])-1:
            y_space_temp -= v_lines_len[i][j+1]
```

```python
mark_temp = gf.Component()
# marker
marker_1 = gf.components.rectangle(size=(4, 1000), layer=(1, 0))
marker_2 = gf.components.rectangle(size=(1000, 4), layer=(1, 0))
marker_basic = gf.Component()
marker_1_ref = marker_basic << marker_1
marker_2_ref = marker_basic << marker_2
marker_1_ref.move((-2, -500))
marker_2_ref.move((-500, -2))
marker_tri = gf.Component()
marker_tri.add_ref(marker_basic).movex(-1000)
marker_tri.add_ref(marker_basic).movey(1000)
marker_tri.add_ref(marker_basic).movey(-1000)
mark_left = mark_temp << marker_tri
mark_left.move((-40700, 0))
mark_right = mark_temp << marker_tri
mark_right.dmirror_x(0).movex(40700)
# mark position
''''
mark1: (-41700, 0), (-40700, 1000), (-40700, -1000)
mark2: (41700, 0), (40700, 1000), (40700, -1000)
'''''
# left label
label_left1 = gf.components.text("(-41700, 0)", size=50, layer=(1, 0))
label_left2 = gf.components.text("(-40700, 1000)", size=50, layer=(1, 0))
label_left3 = gf.components.text("(-40700, -1000)", size=50, layer=(1, 0))
label_left = gf.Component()
label_left.add_ref(label_left1).move((-200, 600))
label_left.add_ref(label_left2).move((700, 1600))
label_left.add_ref(label_left3).move((700, -400))
label_left_ref = mark_temp << label_left
label_left_ref.move((-41700, 0))
# right label
label_right1 = gf.components.text("(41700, 0)", size=50, layer=(1, 0))
label_right2 = gf.components.text("(40700, 1000)", size=50, layer=(1, 0))
label_right3 = gf.components.text("(40700, -1000)", size=50, layer=(1, 0))
label_right = gf.Component()
label_right.add_ref(label_right1).move((-200, 600))
label_right.add_ref(label_right2).move((-1200, 1600))
label_right.add_ref(label_right3).move((-1200, -400))
label_right_ref = mark_temp << label_right
label_right_ref.move((41700, 0))

mark2 = gf.boolean(A = mark_temp, B = mark_temp,  operation="not", layer1=(1, 0), layer2=(10, 0), layer=(2, 0))
mark3 = gf.boolean(A = mark_temp, B = mark_temp,  operation="not", layer1=(1, 0), layer2=(10, 0), layer=(3, 0))
wafer << mark3
wafer << mark2
wafer << mark_temp


# ATTENTION
# mirror the backside pattern for backside photolithography
wafer_backside = gf.boolean(A = wafer, B = wafer,  operation="not", layer1=(3, 0), layer2=(50, 0), layer=(4, 0))
wafer.add_ref(wafer_backside).dmirror_x(0)
wafer.show()
suffix = datetime.now().strftime("%Y%m%d_%H%M")
# wafer.write_gds(f"wafer_{suffix}.gds", write_context_info=False)
wafer.write_gds(f"wafer_{suffix}.gds")
```
